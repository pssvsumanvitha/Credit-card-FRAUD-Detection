```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# distribution of normal transactions and fraud transactions
# here the amount of normal transaction rate is very high
# and fraud transaction rate is low
credit_card_data['Class'].value_counts()
```

```python
# this dataset is highly unbalanced
# 0 --> represents normal transaction
# 1 --> represents fraud transactions
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# comparing the values between legit and fraud
credit_card_data.groupby('Class').mean()
```

```python
#under.sampling. dataset contaiing similar distribution of normal trans and fraud transaction
# no of fraud transations is ---> 492
legit_sample = legit.sample(n = 492)
```

```python
# concatenating two data frames
#axis = 0 ---> row
#axis = 1 ---> column
new_dataset = pd.concat([legit_sample,fraud],axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# from the above result there is still some issue it didnt changed much from before
# if we get a bad sample the mean value will be very different
```

```python
X = new_dataset.drop(columns = 'Class',axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
# spliting the data into training data and testing data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify = Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model = LogisticRegression()
```

```python
#training and logistic Regression Model woth training data
#x_train contains all the features of training data
#y_train contains the corresponding variables
model.fit(X_train,Y_train)
```

```python
print(model)
```

```python
model.fit(X_train,Y_train)
```

```python
#model evaluation
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on Training data:',training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy score on Test Data:',test_data_accuracy)
```
